# Lecture 18 - Natural Language Processing: LLMs, LLM Prompting

**Великі мовні моделі (LLM)** - це тип штучного інтелекту (ШІ), який використовується для обробки природної мови (НЛП). Вони навчаються на величезних масивах даних тексту та коду, що дозволяє їм виконувати широкий спектр завдань, включаючи:

* **Розуміння тексту:** LLM можуть розуміти значення тексту, навіть якщо він складний або неструктурований. Наприклад, вони можуть відповідати на запитання про текст, перекладати мови та виявляти та виправляти помилки в тексті.

* **Генерування тексту:** LLM можуть генерувати текст, який є граматично правильним і логічним. Наприклад, вони можуть створювати різні творчі текстові формати, такі як вірші, код, сценарії, музичні твори, електронні листи, листи тощо.

* **Аналіз тексту:** LLM можуть аналізувати текст, щоб отримати інформацію про його зміст, структуру та стилістику. Наприклад, вони можуть визначати теми тексту, виявляти ключові слова та знаходити зв'язки між різними частинами тексту.

LLM є відносно новим напрямком досліджень ШІ, але вони вже мають широкий спектр застосувань. Вони використовуються в таких сферах, як:

* **Комунікація:** LLM використовуються для створення чат-ботів, які можуть спілкуватися з людьми природним чином.

* **Освіта:** LLM використовуються для створення персоналізованих навчальних програм та надання допомоги студентам у вивченні нових предметів. Наприклад, LLM використовуються для створення підручників, завдань та тестів.

* **Розробка програмного забезпечення:** LLM використовуються для тестування програмного забезпечення, виявлення помилок та генерування коду. Наприклад, LLM використовуються для автоматизації рутинних завдань, таких як написання тестових сценаріїв та рефакторинг коду.

* **Реклама:** LLM використовуються для таргетингу реклами на основі інтересів користувачів. Наприклад, LLM використовуються для створення персоналізованих рекламних оголошень та рекомендацій.

**Як працюють LLM?**

LLM є нейронними мережами, які навчаються на величезних масивах даних тексту та коду. Ці дані використовуються для навчання моделі розпізнавати шаблони в мові. Після навчання LLM можуть використовувати ці шаблони для виконання різних завдань, пов'язаних з обробкою природної мови.

Процес навчання LLM можна поділити на два етапи:

1. **Етап підготовки даних:** На цьому етапі дані очищаються від помилок та готуються до навчання. Цей процес може бути трудомістким, оскільки він вимагає від людини ретельно перевірити дані на наявність помилок.

2. **Етап навчання:** На цьому етапі модель навчається розпізнавати шаблони в даних. Цей процес може бути тривалим, оскільки він вимагає від моделі обробки великої кількості даних.

**Які переваги та недоліки LLM?**

LLM мають ряд переваг, включаючи:

* **Здатність розуміти складний та неструктурований текст**

LLM можуть розуміти текст, навіть якщо він містить неявні або складні зв'язки. Наприклад, LLM можуть розуміти жарти, іронію та метафори.

* **Здатність генерувати текст, який є граматично правильним і логічним**

LLM можуть генерувати текст, який є граматично правильним і логічним, навіть якщо він містить незвичні слова або конструкції. Наприклад, LLM можуть генерувати вірші, код, сценарії та інші творчі текстові формати.

* **Здатність аналізувати текст для отримання інформації про його зміст, структуру та стилістику**

LLM можуть аналізувати текст, щоб отримати інформацію про його зміст, структуру та стилістику. Наприклад, LLM можуть визначати теми тексту, виявляти ключові слова та знаходити зв'язки між різними частинами тексту.

Однак LLM також мають ряд недоліків, включаючи:

* **Можуть бути сприйнятливими до упереджень, які містяться в даних, на яких вони навчаються**

Якщо дані, на яких навчається LLM, містять упередження, то LLM також буде мати ці упередження.

* **Можуть генерувати текст, який є некоректним або шкідливим**

LLM можуть генерувати текст, який є некоректним або шкідливим. Наприклад, LLM можуть генерувати текст, який містить неправдиву інформацію або дискримінаційну мову.

* **Можуть бути дорогими в розробці та обслуговуванні**

Розробка та навчання LLM вимагає значних витрат на обчислювальні ресурси. Крім того, LLM потребують постійного оновлення, оскільки мова постійно змінюється.

**Напрямки досліджень**

Вчені активно працюють над усуненням недоліків LLM. Одним із напрямків досліджень є розробка методів, які дозволяють LLM виявляти та обмежувати упередження, які містяться в даних. Іншим напрямком досліджень є розробка методів, які дозволяють LLM генерувати текст, який є точним і неупередженим.

**Перспективи розвитку**

LLM мають значний потенціал для зміни багатьох аспектів нашого життя. Вони можуть бути використані для створення нових і інноваційних продуктів і послуг, а також для підвищення ефективності та продуктивності в різних сферах.

**Ось деякі приклади того, як LLM можуть бути використані в майбутньому**:

- LLM можуть бути використані для створення персоналізованих освітніх програм, які відповідають індивідуальним потребам кожного студента.
- LLM можуть бути використані для створення нових форм творчого контенту, таких як вірші, музика та художні твори.
- LLM можуть бути використані для автоматизації рутинних завдань, таких як написання звітів і листів.
- LLM можуть бути використані для підвищення ефективності спілкування між людьми та машинами.
 
LLM є потужним інструментом, який має потенціал змінити наше життя. Однак важливо усвідомлювати обмеження LLM і розробляти методи, які дозволяють використовувати їх безпечно та відповідально.

# Prompt Engineering

LLM prompting - це процес використання природномовних інструкцій для інформування та керування поведінкою великої мовної моделі (LLM). LLM - це тип штучного інтелекту, який навчений на величезному наборі даних тексту та коду. Вони можуть генерувати текст, перекладати мови, писати різні види творчого контенту та відповідати на ваші запитання в інформативний спосіб. Однак LLM можуть бути складними для використання. Вони часто не розуміють, що ви від них хочете, і можуть генерувати неправильні або непотрібні результати.

LLM prompting використовується для вирішення цієї проблеми. Він дозволяє вам чітко вказати LLM, що ви хочете, щоб він зробив. Ви можете використовувати prompts для інструктування LLM щодо таких завдань, як:

* **Генерація тексту:** Ви можете використовувати LLM prompting для генерації різних типів тексту, таких як вірші, код, сценарії, музичні твори, електронні листи тощо. Наприклад, ви можете сказати: "Напишіть вірш про кохання" або "Напишіть сценарій для фільму про супергероїв".
* **Переклад мов:** Ви можете використовувати LLM prompting для перекладу текстів з однієї мови на іншу. Наприклад, ви можете сказати: "Перекладіть цей текст з англійської на українську" або "Напишіть лист англійською мовою".
* **Написання різних видів творчого контенту:** Ви можете використовувати LLM prompting для написання різних видів творчого контенту, таких як пісні, оповідання, вірші, сценарії тощо. Наприклад, ви можете сказати: "Напишіть пісню про любов" або "Напишіть оповідання про пригоди".
* **Відповіді на ваші запитання:** Ви можете використовувати LLM prompting для отримання відповідей на ваші запитання. Наприклад, ви можете сказати: "Яка столиця України?" або "Який сьогодні день?".

**Основні принципи LLM prompting**

LLM prompting заснований на наступних принципах:

* **LLM розуміють текст.** Ви повинні використовувати текст для інструктування LLM.
* **LLM навчаються на прикладах.** Чим більше прикладів ви подасте LLM, тим краще він буде розуміти ваші інструкціі.
* **LLM недосконалі.** Вони іноді роблять помилки. Ви повинні бути готові до цього.

**Типи LLM prompts**

Існує багато різних типів LLM prompts. Найпоширеніші з них включають:

* **Описові prompts:** Ці prompts описують те, що ви хочете, щоб LLM зробив. Наприклад, ви можете сказати: "Напишіть вірш про любов" або "Перекладіть цей текст з англійської на українську".
* **Інструкції:** Ці prompts інструктують LLM, що робити. Наприклад, ви можете сказати: "Закінчіть це речення" або "Дай відповідь на це запитання".
* **Запитання:** Ці prompts ставлять запитання LLM. Наприклад, ви можете сказати: "Яка столиця України?".

**Як написати ефективний LLM prompt**

Щоб написати ефективний LLM prompt, дотримуйтесь таких порад:

* **Будьте конкретними.** Чим конкретнішим буде ваш prompt, тим краще LLM його зрозуміє. Наприклад, замість того, щоб сказати "Напишіть вірш", скажіть "Напишіть вірш про любов".
* **Використовуйте просту мову.** LLM не розуміють складної мови. Наприклад, замість того, щоб сказати "Напишіть вірш про кохання, який буде відображати складність людських почуттів", скажіть "Напишіть вірш про кохання".
* **Надавайте приклади.** Якщо ви можете надати LLM приклад того, що ви хочете, щоб він зробив, це допоможе йому краще зрозуміти ваші інструкціі. Наприклад, якщо ви хочете, щоб LLM написав вірш про кохання, ви можете надати йому приклад вірша про кохання, який вам подобається.

Prompt engineering є відносно новою дисципліною для розробки та оптимізації підказок (prompts) для ефективного використання мовних моделей (LM) для широкого спектру програм і дослідницьких тем. Prompt інженерні навички допомагають краще зрозуміти можливості та обмеження великих мовних моделей (LLM).

Дослідники використовують prompt engineering, щоб покращити спроможність LLMs для вирішення широкого кола звичайних і складних завдань, таких як відповіді на запитання та арифметичні задачі. Розробники використовують prompt engineering для розробки надійних і ефективних методів підказок, які взаємодіють з LLM та іншими інструментами.

Prompt engineering — це не лише проектування та розробка підказок. Він охоплює широкий спектр навичок і технік, корисних для взаємодії та розвитку з LLM. Це важливий навик для взаємодії, створення та розуміння можливостей LLM. Ви можете використовувати prompt engineering , щоб підвищити безпеку LLM і створити нові можливості, як-от доповнити LLM знаннями домену та зовнішніми інструментами.

## Prompt formatting

### Zero-shot and Few-shot prompting

Стандартний підхід:

```
<Question>?
```

або 

```
<Instruction>
```

Ви можете відформатувати це у формат відповіді на питання (QA), який є стандартним для багатьох наборів даних QA, як показано нижче:

```
Q: <Question>?
A:
```

Підказка, подібна до наведеної вище, також називається ***zero-shot***, тобто ви безпосередньо пропонуєте моделі відповісти без будь-яких прикладів чи демонстрацій щодо завдання, яке ви хочете виконати. Деякі великі мовні моделі дійсно мають можливість виконувати zero-shot підказки, але це залежить від складності та знання завдання.

Враховуючи наведений вище стандартний формат, одна популярна й ефективна техніка підказок називається ***few-shot***, де ви надаєте приклади (тобто демонстрації). Ви можете відформатувати підказки з кількома випадками таким чином:

```
<Question>?
<Answer>
<Question>?
<Answer>
<Question>?
<Answer>
<Question>?
```

Версія у форматі QA виглядатиме так:

```
Q: <Question>?
A: <Answer>
Q: <Question>?
A: <Answer>
Q: <Question>?
A: <Answer>
Q: <Question>?
A:
```

Майте на увазі, що використовувати формат QA не потрібно. Формат підказки залежить від поставленого завдання. Наприклад, ви можете виконати просте завдання класифікації та навести приклади, які демонструють це завдання, як показано нижче:

*Prompt:*

```
This is awesome! // Positive
This is bad! // Negative
Wow that movie was rad! // Positive
What a horrible show! //
```

*Output:*

```
Negative
```

## General Tips for Designing Prompts

### The Instruction

Ви можете створювати ефективні підказки для різних простих завдань, використовуючи команди, щоб вказати моделі, чого ви хочете досягти, наприклад «Написати», «Класифікувати», «Підсумувати», «Перекласти», «Упорядкувати» тощо.

Майте на увазі, що вам також потрібно багато експериментувати, щоб побачити, що працює найкраще. Спробуйте різні інструкції з різними ключовими словами, контекстами та даними та подивіться, що найкраще підходить для вашого конкретного випадку використання та завдання. Зазвичай, чим конкретнішим і релевантнішим є контекст для завдання, яке ви намагаєтеся виконати, тим краще.

Інші рекомендують розміщувати інструкції на початку підказки. Дана рекомендація полягає в тому, щоб використовувати якийсь чіткий роздільник, наприклад "###", щоб розділити інструкцію та контекст.

Наприклад:

*Prompt:*

```
### Instruction ###
Translate the text below to Spanish:
Text: "hello!"
```

*Output:*

```
¡Hola!
```

*Code Prompt:*

```
### Instruction ###
Refactor code:
Code:
`
import torch

...
`

Output:

```

### Specificity

Будьте дуже конкретні щодо інструкцій і завдань, які ви хочете виконати моделлю. Чим описовішим і детальнішим є підказка, тим кращі результати. Це особливо важливо, коли у вас є бажаний результат або стиль, якого ви прагнете. Немає конкретних маркерів або ключових слів, які дають кращі результати. Важливіше мати хороший формат і описову підказку. Фактично, надання прикладів у підказці є дуже ефективним для отримання бажаного результату в певних форматах.

Розробляючи підказки, ви також повинні пам’ятати про довжину підказки, оскільки існують обмеження щодо довжини підказки. Думайте про те, наскільки конкретними та детальними ви повинні бути. Включення занадто великої кількості непотрібних деталей не обов’язково є хорошим підходом. Деталі мають бути актуальними та сприяти виконанню поставленого завдання. Це те, з чим вам доведеться багато експериментувати.

Як приклад, давайте спробуємо просту підказку витягти певну інформацію з фрагмента тексту.

*Prompt:*

```
Extract the name of places in the following text. 
Desired format:
Place: <comma_separated_list_of_company_names>
Input: "Although these developments are encouraging to researchers, much is still a mystery. “We often have a black box between the brain and the effect we see in the periphery,” says Henrique Veiga-Fernandes, a neuroimmunologist at the Champalimaud Centre for the Unknown in Lisbon. “If we want to use it in the therapeutic context, we actually need to understand the mechanism.""
```

*Output:*

```
Place: Champalimaud Centre for the Unknown, Lisbon
```

### Avoid Impreciseness

Враховуючи наведені вище поради щодо деталізації та вдосконалення формату, можна легко потрапити в пастку бажання бути занадто точним щодо підказок і потенційно створити неточні описи. Часто краще бути конкретним і прямим. Аналогія тут дуже схожа на ефективну комунікацію: чим пряміше, тим ефективніше буде повідомлення.

Наприклад, вам може бути цікаво вивчити концепцію prompt engineering. Ви можете спробувати щось на кшталт:

```
Explain the concept prompt engineering. Keep the explanation short, only a few sentences, and don't be too descriptive.
```

### Text Classification

Досі ви використовували прості інструкції для виконання завдання. Як prompt інженер, ви повинні стати кращими в наданні кращих інструкцій. Але це ще не все! Ви також побачите, що для складніших випадків використання просто надати інструкції буде недостатньо. Тут вам потрібно більше подумати про контекст і різні елементи, які можна використовувати в підказці. Інші елементи, які ви можете надати, це `input data` або `examples`.


*Prompt:*

```
Classify the text into neutral, negative or positive. 
Text: I think the food was okay. 
Sentiment:
```

*Output:*

```
Neutral
```

Ви дали вказівку класифікувати текст, і модель відповіла `'Neutral'`, що правильно. У цьому немає нічого поганого, але припустімо, що вам справді потрібно, щоб модель надала мітку саме в потрібному форматі. Тож замість `Neutral` ви хочете, щоб він повертав `neutral`. Як цього досягти? Існують різні способи зробити це. Тут вам важлива конкретність, тож чим більше інформації ви можете надати, тим кращі результати. Ви можете спробувати навести приклади, щоб визначити правильну поведінку. Давайте спробуємо ще раз:


*Prompt:*

```
Classify the text into neutral, negative or positive. 
Text: I think the vacation is okay.
Sentiment: neutral 
Text: I think the food was okay. 
Sentiment:
```

*Output:*

```
neutral
```

Ідеально! Цього разу модель повернула `neutral`, що є конкретною міткою, яку ви шукали. Схоже, що приклад, наданий у підказці, допоміг моделі бути конкретними у своїх результатах.

Щоб підкреслити, чому іноді важливо бути конкретним, перегляньте приклад нижче та знайдіть проблему:

*Prompt:*

```
Classify the text into nutral, negative or positive. 
Text: I think the vacation is okay.
Sentiment:
```

*Output:*

```
Neutral
```

У чому тут проблема? Як натяк, модель повністю ігнорує вигаданий ярлик `nutral`. Натомість модель видає `Neutral`, оскільки має деяку упередженість щодо цієї мітки. Але припустімо, що ви справді хочете `nutral`. Як би ви це виправили? Можливо, ви можете спробувати додати описи до міток або додати більше прикладів до підказки?
